```python
xml_list = []

for i in range(len(lines)):
# for i in range(2):
    curr_ann = lines[i]
    
    im_name = curr_ann['filename'][17:]
    im = io.imread('./sim_training_data_large/sim_data_capture/' + im_name)
    shape_tuple = tuple([im.shape[1], im.shape[0]])

    len_ann = len(curr_ann['annotations'])
#     print(curr_ann['annotations'])
    if len_ann > 0:
        for j in range(len_ann):
            class_name = curr_ann['annotations'][j]['class']
            xmin      = int(round(max(0, curr_ann['annotations'][j]['xmin'])))
            ymin      = int(round(max(0, curr_ann['annotations'][j]['ymin'])))
            xmax      = int(round(curr_ann['annotations'][j]['x_width'] + xmin))
            ymax      = int(round(curr_ann['annotations'][j]['y_height'] + ymin))
            
            value = (im_name,) + shape_tuple + (class_name,) + tuple([xmin, ymin, xmax, ymax])
            xml_list.append(value)
            
#     else:  #to add 4th label      
#         class_name = 'Other'
#         xmin      = 0
#         ymin      = 0
#         xmax      = im.shape[1]
#         ymax      = im.shape[0]

#         value = (im_name,) + shape_tuple + (class_name,) + tuple([xmin, ymin, xmax, ymax])
#         xml_list.append(value)

#     class_name = 'stop'
#     x = tuple([int(i) for i in lines[i].split()])
#     value = (im_name,) + shape_tuple + (class_name,) + tuple(x)
#     xml_list.append(value)
```
